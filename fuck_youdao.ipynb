```python
import requests
img_save_path = './imgs/'#文件保存路径（可以自己改，默认保存到脚本的同级目录）

def down_pic(down_url, picname):
    header = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.13; rv:62.0) Gecko/20100101 Firefox/62.0'
    }
    req = requests.get(url=down_url, headers=header, cookies=cookie)
    req.encoding = 'utf-8'
    with open('%s%s.jpeg' %(img_save_path,picname), "wb") as f:  # 开始写文件，wb代表写二进制文件 图片已数字命名
        f.write(req.content)
```

```python
def main():
    picname = 0
    for url in res:
        print('第%s个文件长在下载'%picname)
        picname += 1
        down_pic(url, picname)
```
